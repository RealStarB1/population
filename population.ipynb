```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
countries=pd.read_csv('D:/B1/work/BigData/Exam/Dataset/countries.csv')
```

```python
countries.head()
```

```python
data_1997=countries[countries['year']==1997]
```

```python
top_10_p=data_1997.sort_values('population',ascending=False).head(10)
```

```python
plt.figure(figsize=(12,6))

x=range(10)
teamColours = ['#034694','#001C58','#5CBFEB','#D00027',
              '#EF0107','#DA020E','#274488','#ED1A3B',
               '#000000','#091453']
plt.bar(x,top_10_p['population']/10**9,color = teamColours)
plt.xticks(x,top_10_p['country'])
plt.xlabel('Countries')
plt.ylabel('Population in Billions')
plt.title('10 Most Populous Countries in 1997')

plt.show()
```

```python
teamColours = ['#034694','#001C58','#5CBFEB','#00877B',
              '#EF0107','#DA020E','#FF0C9F','#F7FB05',
               '#000000','#091462']
plt.figure(figsize=(12,7))


y=range(10)
plt.barh(y,top_10_p['gdpPerCapita'],color = teamColours)
plt.yticks(y,top_10_p['country'])


plt.xlabel('gdpPerCapita')
plt.title('Top10 - gdpPerCapita in 1997')

plt.show()
```
